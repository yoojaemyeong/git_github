# git, github 고급과정  
## 브랜치를 사용하는 이유
| <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 상황 </span> | <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 이유 </span> |
| --- | --- |
| <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 새로운 기능 추가 </span> | <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 안정적인 버전의 프로젝트에서 새로운 기능을 추가할때 [main] 브랜치의 최신 커밋으로부터 브랜치를 생성해서 개발 </span> |  
| <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 버그 수정 </span> | <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 오류와 버그는 항상 있기 마련이고, 버그가 발생하면 [main]브랜치로부터 새로운 브랜치를 생성해서 작업<br>이때 브랜치 이름은 [hotfix] 또는 [bugfix] 같은 이름을 사용 </span> |  
| <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 병합과 리베이스 테스트 </span> | <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 병합과 리베이스는 입문자에게 가장 까다로운 일 중 하나<br>이럴 때 임시 브랜치를 만들어서 병합과 리베이스 테스트 진행 후 [main] 브랜치로 이동하여 병합<br>잘못되었을 경우 그냥 브랜치를 삭제하면 끝 </span> |  
| <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 이전 코드 개선 </span> | <span style="color:blue;background-color:white;font-size:18px;font-weight:bold;"> 이미 기능 구현은 완료되었는데 코드가 마음에 들지 않아서 함수의 로직 등을 개선하고 싶을 때<br>대부분의 개발자는 이런 상황에서 기존 코드를 주석처리하고 그 아래에 새로운 개선 코드를 작성한느데, 이 방법보다는 브랜치 사용을 추천 </span> |

## 브랜치 관련 명령어
<span style="color:blue;background-color:white;border-bottom:1px dotted #505050;font-size:18px;font-weight:bold;"> `git branch [-v]` 로컬 저장소의 브랜치 목록 보기, -v 옵션을 사용하면 마지막 커밋도 함께 표시, 표시된 브랜치 중 왼쪽에 *가 붙어 있으면 HEAD 브랜치<br>
`git branch [-f] <branch name> [커밋 체크섬]` 새로운 브랜치를 생성, [커밋 체크섬] 값을 주지 않으면 HEAD로부터 브랜치 생성<br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-f 옵션은 이미 있는 브랜치를 다른 커밋으로 옮기고 싶을때 사용<br>
`git branch -r[v]` 원격저장소에 있는 브랜치를 보고 싶을 때 사용, -v 옵션을 추가하면 커밋 요약 추가 보기<br>
`git switch <브랜치 이름>` 특정 브랜치로 변경<br>
`git merge <대상 브랜치>` 현재 브랜치와 대상 브랜치를 병합할 때 사용<br>
`git rebase <대상 브랜치>` 현재 브랜치의 커밋들을 대상 브랜치에 재배치<br>
`git branch -d <브랜치 이름>` 특정 브랜치 삭제, 단 HEAD 브랜치나 병합되지 않는 브랜치는 삭제 불가<br>
`git branch -D <브랜치 이름>` 브랜치 강제 삭제, 단 HEAD 브랜치나 병합되지 않는 브랜치는 삭제 불가<br>
</span>
