# git, github 활용방법

1. 환경구성
   * 소스버전관리시스템: [Git](https://git-scm.com/downloads) 
   * Git 호스팅 사이트 종류: [Github.com](https://github.com), [Gitlab.com](https://gitlab.com), [BitBucket.org](https://bitbucket.org) 
   * 윈도우,리눅스,MacPC 등에 <span style="color:blue;font-weight:bold;">Git 소스버전관리시스템</span> 설치하고 로컬저장소에서 커밋 관리하기 
2. 내 컴퓨터에 Git 설치하기
   * [Git 최신버전 다운로드](https://git-scm.com/downloads) 후 설치
   * 설치된 도구 중 <span style="color:blue;font-weight:bold;">Git Bash</span> 실행
   * 프롬프트창에서 "git" 입력후 엔터하면 git 사용법 등이 출력되면 정상적으로 설치되었음 
3. Git 명령어를 이용하여 로컬저장소 만들기
   * 로컬저장소로 사용하고자 하는 폴더로 이동( cd  )
   * git init <엔터> 
   * 로컬저장소 폴더내에 <span style="color:blue;font-weight:bold;">.git</span>폴더가 생성됨<br>
   
4. Github 원격저장소에 new repository 만들기
   * 원격저장소 만들기(참고로, Github에서 원격저장소를 레포지토리repository라 지칭함.<br>
    new repository "git_github" 생성
   * new repository "git_github"가 생성하면 아래와 같이 "Quick setup" 가이드를 보여줌<br>
   
5. new repository에 커밋 올리기
   * 접근 방법
     

* remote add origin https://github.com/~~~.git
* git push -u origin main :: 원격저장소에 로그인하는 단계로 생성된 토큰ID 및 토큰 입력
